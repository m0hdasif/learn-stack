```python
# Day 1: Python Basics

# 1. Variables and Data Types
age = 25  # Integer
height = 5.9  # Float
name = "John"  # String
fruits = ["apple", "banana", "cherry"]  # List
person = {"name": "Alice", "age": 30, "city": "New York"}  # Dictionary

# 2. Print function
print("Hello, world!")  # Prints the text
print(age)  # Prints the value of the variable 'age'

# 3. Arithmetic Operators
sum_result = 10 + 5  # 15
diff_result = 10 - 4  # 6
product = 4 * 5  # 20
division = 8 / 2  # 4.0
floor_div = 7 // 2  # 3
remainder = 7 % 2  # 1
power = 2 ** 3  # 8
print(sum_result, diff_result, product, division, floor_div, remainder, power)

# 4. Comparison Operators
print(5 == 5)  # True
print(5 != 3)  # True
print(5 > 3)  # True
print(5 < 10)  # True
print(5 >= 3)  # True
print(5 <= 6)  # True

# 5. Logical Operators
print((5 > 3) and (8 > 2))  # True
print((5 > 3) or (3 > 7))  # True
print(not(5 > 3))  # False

# 6. String Operations
greeting = "Hello, " + "John!"  # String Concatenation
print(greeting)

# String Slicing
word = "Hello"
substring = word[1:4]  # Extracts 'ell'
print(substring)

# String Repetition
repeated = "Hi" * 3  # "HiHiHi"
print(repeated)

# 7. Type Casting
int_num = 5
float_num = float(int_num)  # Convert int to float
print(float_num)

num_str = "100"
num = int(num_str)  # Convert string to integer
print(num)

# 8. Comments
# This is a comment. It won't affect the code execution.
# Here is another comment explaining the code.

# 9. Taking Input
name_input = input("Enter your name: ")  # Prompt for name
print("Hello, " + name_input)  # Output: Hello, <name>

# 10. Lists
fruits = ["apple", "banana", "cherry"]
fruits.append("orange")  # Add to list
print(fruits)

fruits.remove("banana")  # Remove from list
print(fruits)

# 11. Tuples
colors = ("red", "green", "blue")
print(colors)

# 12. Dictionaries
person = {"name": "Alice", "age": 30, "city": "New York"}
print(person["name"])  # Access a value by key

# 13. Sets
numbers = {1, 2, 3, 4, 5}
numbers.add(6)  # Add item
print(numbers)

numbers.remove(3)  # Remove item
print(numbers)

# 14. Boolean Values
is_true = True
is_false = False
print(is_true and is_false)  # False
print(is_true or is_false)  # True
print(not is_true)  # False

# 15. Escape Characters
print("Hello\nWorld")  # Newline character
print("Hello\tWorld")  # Tab character
print("C:\\Users\\Name")  # Escape backslash

# 16. Indentation in Python
if 5 > 3:
    print("Five is greater than three.")
else:
    print("Five is not greater than three.")
```
